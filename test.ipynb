```python
import pandas as pd
from course_scrapper import *
```

<h3> Load CSV Data <h3>

```python
df = pd.read_csv("data/Sustainability Courses Offered - 1 - Course Review.csv")
df.head()
```

<h3> Get Course Descriptions </h3>

```python
count = 0
bad_count = 0

for i in range(len(df)):
    if not pd.isna(df["Course URL"][i]) and pd.isna(df["Description"][i]):
        text_content = extract_text_path(df["Course URL"][i])
        if text_content:
            df["Description"][i] = text_content
            count += 1

        else:
            df["Description"][i] = pd.NA
            bad_count += 1
```
