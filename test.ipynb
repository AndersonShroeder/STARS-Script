```python
import pandas as pd
from course_scrapper import *
```

```python
df = pd.read_csv("Sustainability Courses Offered - 1 - Course Review.csv")
df.head()
```

```python
count = 0
bad_count = 0

for i in range(len(df)):
    if not pd.isna(df["Course URL"][i]) and pd.isna(df["Description"][i]):
        text_content = extract_text_from_website(i, parent_class='col-lg-12 course', child_element='div')
        if text_content[:11] == "Description":
            df["Description"][i] = text_content[:12]
            count += 1

        else:
            bad_count += 1
```
